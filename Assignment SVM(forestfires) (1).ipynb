```python
X=data.iloc[:,:-1]
Y=data.iloc[:,-1]
```

```python
x=pd.get_dummies(X)
```

```python
x
```

```python
y=Y.replace({'small':0,'large':1})
```

```python
y
```

```python
#model validation technique(train test split)
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=30)
```

### Grid Search CV

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
clf = SVC()  
#clf.get_params().keys()
param_grid = [{'kernel':['rbf'],'gamma':[15,5,0.5,0.005,0.0005,0.00005],'C':[0.2,1,0.5,0.1,0.001,0.2] }]
gsv = GridSearchCV(clf,param_grid,cv=10)
gsv.fit(x_train,y_train)
print(gsv.best_params_)
print(gsv.best_score_ )
```

```python
clf = SVC()  #clf.get_params().keys()
param_grid = [{'kernel':['poly'],'gamma':[15,5,0.5,0.005,0.0005,0.00005],'C':[0.2,1,0.5,0.1,0.001,0.2] }]
gsv = GridSearchCV(clf,param_grid,cv=10)
gsv.fit(x_train,y_train)
print(gsv.best_params_)
print(gsv.best_score_ )
```

```python
clf = SVC()  #clf.get_params().keys()
param_grid = [{'kernel':['sigmoid'],'gamma':[15,5,0.5,0.005,0.0005,0.00005],'C':[0.2,1,0.5,0.1,0.001,0.2] }]
gsv = GridSearchCV(clf,param_grid,cv=10)
gsv.fit(x_train,y_train)
print(gsv.best_params_)
print(gsv.best_score_ )
```

## Using poly Kernel

```python
#Build a model
model1=SVC(kernel='poly',C=0.61,gamma=0.0003) #default degree=3
model1.fit(x_train,y_train)
y_pred=model1.predict(x_test)
acc=accuracy_score(y_test,y_pred)*100
print("Accuracy = ",acc)
confusion_matrix(y_test,y_pred)
```

## Using RBF Kernel

```python
#Build a model
model2=SVC(C=0.75,gamma=0.00031,kernel='rbf')  # default='rbf' kernal  and default C=1 (panalty) 
model2.fit(x_train,y_train)
y_pred=model2.predict(x_test)
acc=accuracy_score(y_test,y_pred)*100
print("Accuracy = ",acc)
confusion_matrix(y_test,y_pred)
```

## Using Sigmoid Kernel

```python
model3=SVC(C= 0.2, gamma= 0.15, kernel= 'sigmoid')
model3.fit(x_train,y_train)
y_pred=model3.predict(x_test)
acc=accuracy_score(y_test,y_pred)*100
print("Accuracy = ",acc)
confusion_matrix(y_test,y_pred)
```
